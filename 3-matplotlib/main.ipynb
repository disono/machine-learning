### Workflow

```python
# import matplotlib and get it ready for plotting in jupyter
# 1. import matplotlib.pyplot as plt

# 2. prepare data
x = [1, 2, 3, 4]
y = [23, 55, 63, 77]

# 3. setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 4. add the data
ax.plot(x, y)

# 5. customize
ax.set(title="Sample Plot", xlabel="x-axis", ylabel="y-axis")

# 6. save
fig.savefig('./workflow.png')
```

### Making figures with Numpy arrays

```python
x = np.linspace(0, 10, 100)
x
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make a plot from dictionary
nut_butter = {"Almonds": 12,
             "Peanut": 53,
             "Cashew": 235}
fig, ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values())
ax.set(title="Nut's Butter", ylabel="Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter.keys()), list(nut_butter.values()));
```

```python
# histograms and plot it
x = np.random.random(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two option for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter.keys(), nut_butter.values())
ax4.hist(np.random.random(1000))
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))

# data
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter.keys(), nut_butter.values());
ax[1, 1].hist(np.random.random(1000));
```
